```python
import pandas as pd

df = pd.read_csv('temperatures.csv')
df.head()
```

```python
# Display non-rolling averages
import plotly.express as px

px.line(df.dropna(), x='year', y=['avg_temp', 'global_avg_temp'], title='Global vs Columbus Temperatures')
```

```python
df['Columbus'] = df['avg_temp'].rolling(7).mean()
df['Global'] = df['global_avg_temp'].rolling(7).mean()
df.tail()
```

```python
# Display rolling averages
px.line(df.dropna(), x='year', y=['Global', 'Columbus'], title='Global vs Columbus Temperatures - 7 Day Rolling Average')
```
